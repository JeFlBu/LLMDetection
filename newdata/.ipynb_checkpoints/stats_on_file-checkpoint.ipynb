```python
import pandas as pd
import matplotlib.pyplot as plt
import json
import numpy as np
import os
```

```python

current_dir = os.getcwd()

# Iterate through each item in the current directory
for item in os.listdir(current_dir):
    item_path = os.path.join(current_dir, item)
    if '.ipynb_checkpoints' in item_path:
        continue
    if not 'strategygenerate/' in item_path:
        continue
    if not os.path.exists(item_path+'/results_scores_email_VARIANT061124_WITHPROB_TEMP2.json'):
        continue
    # Check if it's a directory
    if not os.path.isdir(item_path):
        continue
    print(item_path)
    
    stats_res=dict()
    for file_n in [item_path+'/results_scores_email_VARIANT061124_WITHPROB_TEMP2.json']:
        # if os.path.exists(file_n.replace('.json','')+'stats_res.json'):
        #     print('already done')
        #     continue
        stats_res[file_n]=dict()
        f=open(file_n)
        tmp=json.load(f)
        f.close()
        f=open(file_n.replace('_VARIANT061124_WITHPROB_TEMP2.json','_from_may23_VARIANT061124_WITHPROB_TEMP2.json'))
        tmp['fromMay24']=json.load(f)['res']
        f.close()
        
        tmp2=[]
        for key in tmp.keys():
            tmp2+=tmp[key]
            stats_res[file_n][key]=dict()
            arr = np.array([x[1] for x in tmp[key]])
    
            # Calculating statistics
            mean = np.mean(arr)
            std_dev = np.std(arr)
            minimum = np.min(arr)
            maximum = np.max(arr)
            median = np.median(arr)
            q1 = np.percentile(arr, 25)  # First quartile (25th percentile)
            q3 = np.percentile(arr, 75)  # Third quartile (75th percentile)
            
            stats_res[file_n][key]["Mean"]=mean
            stats_res[file_n][key]["Standard Deviation"]=std_dev
            stats_res[file_n][key]["Min"]=minimum
            stats_res[file_n][key]["Max"]=maximum
            stats_res[file_n][key]["Median"]=median
            stats_res[file_n][key]["First Quartile (Q1)"]= q1
            stats_res[file_n][key]["Third Quartile (Q3)"]= q3
        # Example list of lists with [score, date]
        data = [x[1:] for x in tmp2]
        
        # Convert to DataFrame
        df = pd.DataFrame(data, columns=['score', 'date'])
        
        # Convert the 'date' column to datetime
        df['date'] = pd.to_datetime(df['date'])
        
        # Extract the year from the date
        df['year'] = df['date'].dt.year
        
        # Group by the year and calculate the average score
        df_avg = df.groupby('year')['score'].mean().reset_index()
        
        # Plot the average score by year
        plt.figure(figsize=(8, 6))
        plt.plot(df_avg['year'], df_avg['score'], marker='o')
        plt.title('Average Score by Year for '+file_n.replace('.json',''))
        plt.xlabel('Year')
        plt.ylabel('Average Score')
        plt.grid(True)
        plt.savefig(file_n.replace('.json','')+'_year.png', dpi=300, bbox_inches='tight')
        plt.close()
    
        data =[x[1:] for x in tmp2]
        # Convert to DataFrame
        df = pd.DataFrame(data, columns=['score', 'date'])
        df = df[df['date'].notna()]

        # Convert the 'date' column to datetime
        df['date'] = pd.to_datetime(df['date'])
        
        # Create a new column with year-month (e.g., '2023-02')
        df['year_month'] = df['date'].dt.to_period('M')
        
        # Group by the year-month and calculate the average score
        df_avg_month = df.groupby('year_month')['score'].mean().reset_index()
        
        # Convert 'year_month' back to datetime for plotting purposes
        df_avg_month['year_month'] = df_avg_month['year_month'].dt.to_timestamp()
        
        # Plot the average score by year-month
        plt.figure(figsize=(10, 6))
        plt.plot(df_avg_month['year_month'], df_avg_month['score'], marker='o')
        plt.title('Average Score by Month for '+file_n.replace('.json',''))
        plt.xlabel('Month')
        plt.ylabel('Average Score')
        plt.xticks(rotation=45)
        plt.grid(True)
        # plt.ylim(0.54, 0.555)
        plt.savefig(file_n.replace('.json','')+'_month.png', dpi=300, bbox_inches='tight')
        plt.close()
    
        df['year_month'] = df['year_month'].dt.to_timestamp()  # Convert back to datetime for filtering

        # Loop through each unique month
        for month in df['year_month'].unique():
            # Filter data for the current month
            month_data = df[df['year_month'] == month]['score']
            
            # Plot the normalized frequency distribution of 'score' values for the month
            plt.figure(figsize=(8, 6))
            plt.hist(month_data, bins=40, range=(0, 1), color='lightcoral', edgecolor='black', density=True)
            plt.title(f'Distribution of Scores for {month.strftime("%Y-%m")}')
            plt.xlabel('Score of detection')
            plt.ylabel('Frequency')
            plt.grid(True)
            plt.ylim(0, 10)
            plt.savefig(file_n.replace('.json', '') + f'_{month.strftime("%Y-%m")}_normalized_frequency.png', dpi=300, bbox_inches='tight')
            plt.close()

        
        for threshold in [0.45,0.55,0.65,0.75,0.85,0.93,0.94,0.95,0.96,0.97,0.98,0.99]:
            
            # Convert to DataFrame
            df = pd.DataFrame(data, columns=['score', 'date'])
            df = df[df['date'].notna()]

            # Convert the 'date' column to datetime
            df['date'] = pd.to_datetime(df['date'])
            
            # Create a new column with year-month (e.g., '2023-02')
            df['year_month'] = df['date'].dt.to_period('M')
            
            # Group by year-month and calculate the percentage of values above the threshold
            df_above_threshold = df.groupby('year_month').apply(lambda x: (x['score'] > threshold).mean() * 100).reset_index(name='percentage_above_threshold')
            
            # Convert 'year_month' back to datetime for plotting purposes
            df_above_threshold['year_month'] = df_above_threshold['year_month'].dt.to_timestamp()
            
            # Plot the histogram (as a bar plot)
            plt.figure(figsize=(10, 6))
            plt.bar(df_above_threshold['year_month'], df_above_threshold['percentage_above_threshold'], width=20, color='skyblue')
            plt.title(f'Percentage of Values Above {threshold} by Month for {file_n}')
            plt.xlabel('Month')
            plt.ylabel('Percentage of Values Above Threshold')
            plt.xticks(rotation=45)
            plt.grid(True)
            plt.savefig(file_n.replace('.json','')+'_'+str(threshold)+'_barplot.png', dpi=300, bbox_inches='tight')
            plt.close()
        f=open(file_n.replace('.json','')+'stats_res.json','w')
        json.dump(stats_res,f)
        f.close()
```
